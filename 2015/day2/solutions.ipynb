```python
# Preprocessing

with open("input.txt", "r") as f:
    dimensions: list[tuple[int, int, int]] = []
    while dimension := f.readline():
        length, width, height = dimension.split("x")
        dimensions.append((int(length), int(width), int(height)))
```

```python
# Puzzle 1

import itertools


def wrapping_paper_area(l: int, w: int, h: int) -> int:
    lw_area = l * w
    wh_area = w * h
    lh_area = l * h
    return 2 * sum([lw_area, wh_area, lh_area]) + min(lw_area, wh_area, lh_area)


sum(itertools.starmap(wrapping_paper_area, dimensions))
```

```python
# Puzzle 2

import itertools

def ribbon_area(l: int, w: int, h: int) -> int:
    volume = l * h * w
    lw_perimeter = 2 * (l + w)
    wh_perimeter = 2 * (w + h)
    lh_perimeter = 2 * (l + h)
    return min(lw_perimeter, wh_perimeter, lh_perimeter) + volume

sum(itertools.starmap(ribbon_area, dimensions))
```
