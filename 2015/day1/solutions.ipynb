```python
# Preprocessing

with open("input.txt", "r") as f:
    paths: str = f.read()
```

```python
# Puzzle 1

print(sum(1 if path == "(" else -1 for path in paths))
```

```python
# Puzzle 2

position: int = 0
floor: int = 0
for path in paths:
    position += 1
    if path == "(":
        floor += 1
    else:
        floor -= 1
    if floor == -1:
        break

print(position)
```
