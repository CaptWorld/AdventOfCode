```python
# Proprocessing
input: str = "iwrupvqb"
```

```python
# Utilities

import hashlib


def is_valid(num: int, starts_with: str) -> bool:
    hash = hashlib.md5()
    hash.update(f"{input}{num}".encode())
    return hash.hexdigest().startswith(starts_with)
```

```python
# Puzzle 1

num: int = 0
starts_with: str = "0" * 5
while not is_valid(num, starts_with):
    num += 1

num
```
