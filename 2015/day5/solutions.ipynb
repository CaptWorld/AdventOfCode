```python
# Preprocessing
with open("input.txt", "r") as f:
    strings: list[str] = f.readlines()
```

```python
# Puzzle 1

import re

repeating_chars: re.Pattern = re.compile(r"(\w)\1")
specific_patttern: re.Pattern = re.compile(r"ab|cd|pq|xy")


def is_nice(string: str) -> bool:
    vowels: int = sum(char in "aeiou" for char in string)
    return (
        vowels >= 3
        and bool(repeating_chars.search(string))
        and not specific_patttern.search(string)
    )


sum(is_nice(string) for string in strings)
```
