```python
# Preprocessing
with open("input.txt", "r") as f:
    strings: list[str] = f.readlines()
```

```python
# Puzzle 1

excluded_strings: list[str] = ["ab", "cd", "pq", "xy"]
vowels: set[str] = set("aeiou")


def has_repetition(string: str) -> bool:
    for i in range(0, len(string) - 1):
        if string[i] == string[i + 1]:
            return True
    return False


def has_exclusion_strings(string: str) -> bool:
    return any(excluded_string in string for excluded_string in excluded_strings)


def is_nice(string: str) -> bool:
    num_vowels: int = sum(char in vowels for char in string)
    return (
        num_vowels >= 3 and has_repetition(string) and not has_exclusion_strings(string)
    )


sum(is_nice(string) for string in strings)
```

```python
# Puzzle 2


def has_pairwise_repetition(string: str) -> bool:
    pairs: set[str] = set()
    prev_pair: str = string[0:2]
    for i in range(1, len(string) - 1):
        current_pair: str = string[i : i + 2]
        if current_pair in pairs:
            return True
        pairs.add(prev_pair)
        prev_pair = current_pair
    return False


def has_repeated_enclosing_characters(string: str) -> bool:
    for i in range(0, len(string) - 2):
        if string[i] == string[i + 2]:
            return True
    return False


def is_nice(string: str) -> bool:
    return has_pairwise_repetition(string) and has_repeated_enclosing_characters(string)


sum(is_nice(string) for string in strings)
```
