```python
# Preprocessing

from typing import TypeAlias, Literal

Direction: TypeAlias = Literal[-1, 1]

with open("input.txt", "r") as f:
    rotations: list[tuple[Direction, int]] = []
    while line := f.readline():
        rotations.append(((-1 if line[0] == "L" else 1, int(line[1:]))))
```

```python
# Utilities

TOTAL_POSITIONS: int = 100
```

```python
# Puzzle 1

password: int = 0
current_position: int = 50

for direction, distance in rotations:
    current_position += distance * direction
    if current_position % TOTAL_POSITIONS == 0:
        password += 1

password
```

```python
# Puzzle 2 - Unfinished

password: int = 0
current_position: int = 50


for direction, distance in rotations:
    new_position = current_position + distance * direction

    if current_position % TOTAL_POSITIONS == 0:
        password += distance // TOTAL_POSITIONS
    elif new_position % TOTAL_POSITIONS == 0:
        # 55 -> -100
        password += abs(
            new_position // TOTAL_POSITIONS - current_position // TOTAL_POSITIONS
        )
        if new_position // TOTAL_POSITIONS == current_position // TOTAL_POSITIONS:
            password += 1
    else:
        password += abs(
            new_position // TOTAL_POSITIONS - current_position // TOTAL_POSITIONS
        )

    print(current_position, new_position, password)
    current_position = new_position

password
```
